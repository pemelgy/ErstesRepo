# Verteilung der Gehälter nach Employmenttype und Experiencelevel

```python
# Nur Spalten nehmen
plot_df = df[["salary_usd", "employment_type", "experience_level"]]

plt.figure(figsize=(10, 6))
sns.boxplot(data=plot_df, x="employment_type", y="salary_usd", hue="experience_level", palette="Set2")
plt.xlabel("Anstellungsart")
plt.ylabel("Gehalt USD")
plt.title("Gehaltsverteilung nach Anstellungsart & Erfahrungslevel")
plt.tight_layout()
plt.show()
```

# Histogramm Normalverteilung  - Gehalt und Häufigkeit

```python
plt.figure(figsize=(10, 6))
plt.hist(df["salary_usd"], bins=30, edgecolor='black', alpha=0.7, density=True)
plt.xlabel("Gehalt in USD")
plt.ylabel("Häufigkeit (Dichte)")
plt.title("Verteilung der Gehälter (Histogramm)")
plt.grid(True, alpha=0.3)
plt.show()
```

# Verteilung Benefitsscore

```python
sns.histplot(data=df, x="benefits_score", kde=True, discrete=True)
plt.title("Verteilung Benefits-Score")
plt.show()
```

```python
# Anzahl aller Beobachtungen
n_total = len(df)

# Anzahl der Fälle mit Benefits-Score >= 50
n_ge_50 = (df["benefits_score"] >= 50).sum()

# Anteil als Anteil und als Prozentzahl
anteil_ge_50 = n_ge_50 / n_total
anteil_ge_50_prozent = anteil_ge_50 * 100

print(f"Anzahl >= 50: {n_ge_50} von {n_total}")
print(f"Anteil >= 50: {anteil_ge_50:.2f}")
print(f"Anteil >= 50: {anteil_ge_50_prozent:.1f}%")
```

```python
# Anzahl aller Beobachtungen
n_total = len(df)

# Anzahl der Fälle mit Benefits-Score >= 50
n_ge_50 = (df["benefits_score"] >= 80).sum()

# Anteil als Anteil und als Prozentzahl
anteil_ge_50 = n_ge_50 / n_total
anteil_ge_50_prozent = anteil_ge_50 * 100

print(f"Anzahl >= 50: {n_ge_50} von {n_total}")
print(f"Anteil >= 50: {anteil_ge_50:.2f}")
print(f"Anteil >= 50: {anteil_ge_50_prozent:.1f}%")
```

# Korrelationsmatrix über alle numerischen Werte der Tabelle

```python
spalten = ['salary_usd', 'remote_ratio', 'job_description_length', 
           'benefits_score', 'posting_month', 'posting_year', 'date_diff_days']

# "Zeilennr." explizit ausschließen (falls in df vorhanden)
spalten_numerisch = [s for s in spalten if s != 'Zeilennr.']

korrelation = df[spalten_numerisch].corr()  # Nur gewünschte numerische Spalten

plt.figure(figsize=(8,6))
sns.heatmap(korrelation, annot=True, cmap="coolwarm")
plt.title("Korrelationen zwischen numerischen Variablen der AI Jobs 2024/2025 Tabelle")
plt.show()
```

```python
korrelation = df.select_dtypes(include=['number']).corr()

plt.figure(figsize=(10,8))
sns.heatmap(korrelation, annot=True, cmap="coolwarm", center=0, square=True)
plt.title("Korrelationen numerischer Variablen AI Jobs 2024/2025")
plt.tight_layout()
plt.show()
```

```python
#spalten = ['salary_usd', 'remote_ratio', 'job_description_length', 'benefits_score', 'posting_month', 'posting_year', 'date_diff_days']
           
corr_matrix = df.corr(numeric_only=True)

#korrelation = df[spalten].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Korrelationen zwischen numerischen Variablen der AI Jobs 2024/2025 Tablle")
plt.show()
```

# Anteil Remote-Ratio und Company Size

```python
import pandas as pd

# Gruppiere und berechne Anteile
pivot = pd.pivot_table(
    df,
    index='company_size',
    columns='remote_ratio',
    aggfunc='size',
    fill_value=0
)
pivot = pivot.div(pivot.sum(axis=1), axis=0) * 100  # In Prozent
pivot.columns.name = None
print(pivot.round(1))
```
