# Spalte required_skills splitten an allen Kommas und jeweils für jeden Begriff der Zeile ein neue Splate erstellen

```python
#  df['...'].str.split(',\\s*', expand=True) an: Trennt bei Komma + optionalem Leerzeichen (regex \\s* vermeidet Leerzeichen-Probleme).
# expand=True erzeugt DataFrame mit Spalten 0,1,2,...


split_df = df['required_skills'].str.split(',\\s*', expand=True)
split_df.columns = ['required_skill_1', 'required_skill_2', 'required_skill_3', 'required_skill_4', 'required_skill_5']
df_split = pd.concat([df.drop('required_skills', axis=1), split_df], axis=1)
```
